## Import Packages/Library yang Digunakan

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

## Data Wrangling

### Gathering Data

````python
df = pd.read_csv('/content/Bike Sharing Dataset (day).csv')
df.head()
````

### Assessing Data

````python
df.info()
````

````python
df.isna().sum()
````

````python
print("Jumlah duplikasi: ",df.duplicated().sum())
df.describe()
````

### Cleaning Data

````python
df.drop_duplicates(inplace=True)
print("Jumlah duplikasi: ", df.duplicated().sum())
````

````python
df["dteday"] = pd.to_datetime(df["dteday"])
````

````python
df = df.fillna(0)
````

##Exploratory Data Analysis (EDA)

````python
df.describe(include="all")
````

````python
average_rentals_by_season = df.groupby('season')['cnt'].mean().reset_index()
````

## Visualization & Explanatory Analysis

### Pertanyaan 1: Apakah terdapat perbedaan penggunaan sepeda di musim yang berbeda?

````python
plt.figure(figsize=(8, 6))
sns.barplot(data=average_rentals_by_season, x='season', y='cnt')
plt.title('Rata-Rata Jumlah Peminjaman Sepeda Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Rata-Rata Jumlah Peminjaman')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Spring', 'Summer', 'Fall', 'Winter'])
plt.show()
````

### Pertanyaan 2: Apa pengaruh cuaca terhadap penggunaan sepeda?

````python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='weathersit', y='cnt')
plt.title('Pengaruh Cuaca terhadap Jumlah Peminjaman Sepeda')
plt.xlabel('Cuaca')
plt.ylabel('Jumlah Peminjaman')
plt.xticks(ticks=[0, 1, 2], labels=['Cerah', 'Berawan', 'Hujan/Salju Ringan'])
plt.show()
````

## Conclusion

 Pertanyaan 1: Musim memiliki pengaruh terhadap penggunaan sepeda.Terdapat perbedaan penggunaan sepeda di musim yang berbeda. Rata-rata jumlah peminjaman sepeda lebih tinggi di musim panas dan musim semi dibandingkan dengan musim gugur dan musim dingin.
 Pertanyaan 2: Cuaca memiliki pengaruh yang signifikan terhadap penggunaan sepeda. Cuaca cerah dan hangat mendorong penggunaan sepeda, sedangkan cuaca hujan dan dingin menghambatnya.
